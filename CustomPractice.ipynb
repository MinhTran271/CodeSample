```python
# GCD and LCM calculator
o1 = n1 = int(input('Enter a positive integer: '))
o2 = n2 = int(input('Enter a positive integer: '))

# Set up variables for coefficicents
x1 = 1
y1 = 0
x2 = 0
y2 = 1

while n2 != 0:
    # Compute the coefficients
    x1, x2 = x2, x1 - (n1//n2)*x2
    y1, y2 = y2, y1 - (n1//n2)*y2
    # Compute the gcd
    n1, n2 = n2, n1 % n2

print('Their greatest common divisor is:',n1)
print('Their least common multiple is:',o1*o2/n1)
print(f'We have: {x1}*{o1} + {y1}*{o2} = {x1*o1+y1*o2}')
```
